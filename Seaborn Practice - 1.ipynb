## Seaborn

### Distribution Plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#loading the built-in dataset

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Distribution Plot

```python
#Distribution plot is used for univariate analysis and to plot the distribution of a variable
#KDE --> Kernel Density Estimation
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False, bins = 10) #bins--> number of bins
```

```python
sns.distplot(tips['total_bill'],kde=False, bins = 100) #bins--> number of bins
```

```python
sns.distplot(tips['total_bill'],kde=False, bins = 40) #bins--> number of bins
```

### Joint Plot

```python
#Joint PLot allows us to do bivariate visualizations

sns.jointplot(x ='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x ='total_bill',y='tip',data=tips,kind='hex') #kind --> allows us to make hexagon distribution
```

```python
sns.jointplot(x ='total_bill',y='tip',data=tips,kind='reg')  #reg--> regression
```

```python
sns.jointplot(x ='total_bill',y='tip',data=tips,kind='kde') # kde --> gives the density of the relationship between the two variables
```

### Pair Plot

```python
# Allows us to plot pairwise relationships between all the numerical variables

sns.pairplot(tips)  #Distribution plot is plotted for the variable vs itself
```

```python
sns.pairplot(tips, hue = 'sex') # hue --> passing the categorical variable categorizes the plots as per the variable passed
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

### Rug Plot

```python
#Another way of plotting the distribution plot

sns.rugplot(tips['total_bill'])
```

### KDE Plot

```python
#KDE(Kernel Density Basis Function) --> returns the sum of all the normal distribution plots 

sns.kdeplot(tips['total_bill'])
```

### Categorical Plots

### Bar Plot

```python
#to plot the categorical variable with another categorical variable or numerical variable

sns.barplot(x='sex',y='total_bill',data = tips)  #by group plot

#default estimator --> average (mean) aggregation
```

```python
import numpy as np 

sns.barplot(x='sex',y='total_bill',data = tips, estimator =np.std )   #standard deviation
```

### Count Plot

```python
#Estimator is counting the occurence of the values

sns.countplot(x='sex',data=tips)
```

### Box Plot and Violin Plots --> For plotting the distribution of the categorical data

```python
#Distribution of the total bill per day

sns.boxplot(x='day',y='total_bill',data = tips)
```

```python
#adding another categorical variable
sns.boxplot(x='day',y='total_bill',data = tips, hue = 'smoker')
```

```python
sns.boxplot(x='day',y='total_bill',data = tips, hue = 'sex')
```

### Violin Plots

```python
sns.violinplot(x='day',y='total_bill',data=tips) #Violin plots are similar to box plots

#show the underlying distribution of the data points as well
#little more information in violin plot as compared to box plot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue = 'sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue = 'sex',split=True)
```

### Strip Plot

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter =False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter = True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter = True, hue = 'sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter = True, hue = 'sex', split = True)
```

 ### Swarm Plot --> Combination of strip plot and violin plot

```python
sns.swarmplot(x='day',y='total_bill',data=tips) #shows all the data points (distribution)

#swarm not viable for very large datasets as it shows all the data points
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips, color = 'black')
```

### Factor Plot  --> Generalized form of all the plots shown above

```python
sns.factorplot(x='day',y='total_bill',data=tips, kind = 'bar') #kind --> decides the type of plot we want
```

```python
sns.factorplot(x='day',y='total_bill',data=tips, kind = 'violin')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips, kind = 'swarm')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips, kind = 'box')
```
