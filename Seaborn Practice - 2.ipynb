### Matrix Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
#Heat Map --> Data should already be in matrix format

#Correlation data
tc = tips.corr()
```

```python
tc
```

```python
#Heat Map
sns.heatmap(tc) #correlated data (matrix data) is passed as the parameter for heat map
```

```python
sns.heatmap(tc,annot = True)
```

```python
sns.heatmap(tc,annot = True, cmap = 'coolwarm') #cmap --> color map
```

```python
flights.head()
```

```python
#Converting it to the matrix form

fmat = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
fmat
```

```python
sns.heatmap(fmat)
```

```python
sns.heatmap(fmat, cmap = 'magma')
```

```python
sns.heatmap(fmat, cmap = 'magma', linecolor = 'white',  linewidth = 1)
```

```python
sns.heatmap(fmat, cmap = 'coolwarm', linecolor = 'black',  linewidth = 1)
```

### Cluster Map

```python
sns.clustermap(fmat) #Heirarchical clusters created based on similarities

#Order is not preserved
```

```python
sns.clustermap(fmat, cmap = 'coolwarm') 
```

```python
#Normalizing the scale

sns.clustermap(fmat, cmap = 'coolwarm', standard_scale = 1) 
```

### Regression Plots

```python
tips.head()
```

```python
#Linear Model Plot

sns.lmplot(x='total_bill', y = 'tip', data = tips)
```

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips, hue = 'sex')
```

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o','v'])
```

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o','v'],
          scatter_kws = {'s':100})

#size of the scatters  --> Marker sizing
```

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips, col='sex')
```

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips, col='sex', row = 'time')
```

```python
 sns.lmplot(x='total_bill', y = 'tip', data = tips, col='day', row = 'time', hue = 'sex')
```

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips, col='day', hue = 'sex')
```

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips, col='day', hue = 'sex',
          aspect = 0.6, 
          size = 8)
```
