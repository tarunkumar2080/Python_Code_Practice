```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
#Pair Grid --> ability to custiomize anf gives the control to the user
# Takes all the numerical columns and creates the subplots (grid) for each pair
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)  #mapping the scatter plot to the subplots created by Pair Grid
```

```python
# Plotting different type of plots on the diagonal and upper and lower sides of the diagonal

g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

### Facet Grid

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(sns.distplot, 'tip')
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(plt.scatter,'total_bill', 'tip')
```

### Style and Color

```python
sns.set_style('ticks') #to set the style for future plots

sns.countplot(x='sex',data = tips)
sns.despine() # removes the borders from right and top (lef and bottom can also be removed)
```

```python
plt.figure(figsize = (12,6))
sns.countplot(x='sex',data = tips)
```

```python
plt.figure(figsize = (12,6))
sns.set_context('poster')
sns.countplot(x='sex',data = tips)
```

```python
plt.figure(figsize = (12,6))
sns.set_context('talk')
sns.countplot(x='sex',data = tips)
```
