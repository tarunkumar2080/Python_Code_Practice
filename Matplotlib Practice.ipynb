### Data Visualization - Matplotlib 

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #allows the visulaizations to be seen in jupyter
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional Method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()  #show() is needed to show the plot if the tool used is not jupyter
```

```python
#Putting multi-plots on the same canvas
plt.subplot(1,2,1) #Arguments are number of rows, number of columns, plot numbered being referred
plt.plot(x,y,'r')
#plt.show()

plt.subplot(1,2,2)
plt.plot(y,x,'b')
#plt.show()
```

```python
#Object oriented method
fig = plt.figure()

axes = fig.add_axes([0.5,0.5,0.9,0.9]) #left,bottom,width,height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
#Subplots using object-oriented method
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
```

```python
#Subplots using object-oriented method
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.4,0.4,0.4]) #left, bottom, width, height

axes1.plot(x,y,'g')
axes1.set_title('Larger Plot')

axes2.plot(y,x,'y')
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 3)
plt.tight_layout()
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 3)
axes[0][2].plot(x,y)
axes[1][1].plot(y,x)
plt.tight_layout()
```

### Figsize and DPI

```python
fig = plt.figure(figsize=(3,2),dpi=100) #dpi--> dots per inches (pixels per inch)
#figsize => tuple --> width, height
```

```python
fig = plt.figure(figsize=(8,4)) 

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#Subplots with figzise

fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(8,8))

axes[0][0].plot(x,y)
axes[1][1].plot(y,x)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2, label ='X Squared')
axes.plot(x,x**3, label = 'X Cubed')
axes.legend(loc = 0) #labels are important to have the legends for different plots plotted in the same axes

#loc --> location for the legend

```

### Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', linewidth = 5, alpha = 0.5, linestyle = '--')  #linewidth or lw --> for the line width of the plotline
#alpha --> opacity of the plot line
#color can be RGB Hex code also
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', linewidth = 5, alpha = 0.5, linestyle = ':')  #linewidth or lw --> for the line width of the plotline
#alpha --> opacity of the plot line
#color can be RGB Hex code also
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', linewidth = 5, alpha = 0.5, linestyle = 'steps')  #linewidth or lw --> for the line width of the plotline
#alpha --> opacity of the plot line
#color can be RGB Hex code also
#linestyle or ls for styple of the plot line
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', linewidth = 5, alpha = 0.5, linestyle = '-', marker = 'o', markersize = 20)  #linewidth or lw --> for the line width of the plotline
#alpha --> opacity of the plot line
#color can be RGB Hex code also
#marker to show the data points
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', linewidth = 5, alpha = 0.5, linestyle = '-', marker = '+', markersize = 30)  #linewidth or lw --> for the line width of the plotline
#alpha --> opacity of the plot line
#color can be RGB Hex code also
#marker to show the data points
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', linewidth = 5, alpha = 0.5, linestyle = '-', marker = 'o', markersize = 20, 
       markerfacecolor = 'yellow', markeredgewidth = 3, markeredgecolor = 'green')  #linewidth or lw --> for the line width of the plotline
#alpha --> opacity of the plot line
#color can be RGB Hex code also
#marker to show the data points
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', lw = 3, ls= '--')

ax.set_xlim([0,3])  #limiting the plot on x-axis
ax.set_ylim([0,3]) #limiting on y-axis
```
